```python
import ipywidgets as widgets
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import seaborn as sb
#import ggplot as ggp

#sb.set_palette('RdBu', 16)

class NodeUtilitisationProfile:
    def __init__(self):
        df = pd.read_json('data.json') # 1 experiment
        df2 = pd.read_json('data-set-18-03-2023-1.json') # 2 experiments
        df3 = pd.read_json('data-set-18-03-2023-2.json') # 3 experiments
        self.results = df.data.result
        self.results2 = df2.data.result
        self.results3 = df3.data.result

        self.start = widgets.IntText(description='Start: ')
        self.end = widgets.IntText(description='End: ')
        self.button = widgets.Button(
            description='Click me',
        )
        self.button.on_click(self.node_intensity_between)
        m = widgets.HBox([self.start, self.end])
        display(m)
        display(self.button)

    def node_intensity_between(self, caller):
        start = self.start.get_interact_value()
        end = self.end.get_interact_value()
        slices = {}
        heat_map1 = np.zeros((4,4))
        heat_map2 = np.zeros((4,4))
        heat_map3 = np.zeros((4,4))

        i = 0
        # heat_map[i % 4][int(np.floor(i % 16 / 4.0))]
        for node in self.results:
            heat_map1[int(np.floor(i % 16 / 4.0))][i % 4] = slices[node['metric']['instance']] = np.mean(
                list(map(
                    lambda entry: float(entry[1]),
                    list(filter(lambda entry: int(entry[0]) >= start and int(entry[0]) <= end, node['values']))
                ))
            )
            
            i += 1

        for node in self.results2:
            heat_map2[int(np.floor(i % 16 / 4.0))][i % 4] = slices[node['metric']['instance']] = np.mean(
                list(map(
                    lambda entry: float(entry[1]),
                    list(filter(lambda entry: int(entry[0]) >= start and int(
                        entry[0]) <= end, node['values']))
                ))
            )

            i += 1

        for node in self.results3:
            heat_map3[int(np.floor(i % 16 / 4.0))][i % 4] = slices[node['metric']['instance']] = np.mean(
                list(map(
                    lambda entry: float(entry[1]),
                    list(filter(lambda entry: int(entry[0]) >= start and int(
                        entry[0]) <= end, node['values']))
                ))
            )

            i += 1
        
        slice_frame = pd.DataFrame({'portions': slices.values()})
        slice_frame.plot.pie(y='portions')
        slice_frame.plot.bar()
        plt.figure()
        sb.set_theme(palette='viridis')
        
        p1 = sb.heatmap(data = heat_map1, cmap='RdBu_r', linewidths=0.5, annot=True)

        f, (pa, pb, pc) = plt.subplots(1, 3, figsize=(15, 3))
        pa.set_title("A")
        pb.set_title("B")
        pc.set_title("C")

        sb.heatmap(ax=pa, data=heat_map1, cmap='RdBu_r',
                        linewidths=0.5, annot=True)
        
        sb.heatmap(ax=pb, data=heat_map2, cmap='RdBu_r',
                        linewidths=0.5, annot=True)
        
        sb.heatmap(ax=pc, data=heat_map3, cmap='RdBu_r',
                   linewidths=0.5, annot=True)

        plt.figure()
        #ggp.ggplot(heat_map)

def load_data(a):
    global start, end
    print(start.get_interact_value())
    print(end.get_interact_value())

def panel(start, end):
    print(f"{start} {end}")

run = NodeUtilitisationProfile()


#out = widgets.interactive_output(panel, {'start': start, 'end': end})
```
